```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import os


def process_image(image_input_full):
    #Cropping the image for better performance
    image_input = image_input_full[ 440:680,100:1200]
    
    # point list of the polygonal area on image where we want to apply our filters
    region_coordinates =[[(0,240), (480,0), (600,0), (1100,240)]]  
    
    # Crop the images with polygonal region
    mask = np.zeros_like(image_input)    
    vertices = np.array(region_coordinates, dtype=np.int32)
    cv2.fillPoly(mask, vertices, (255, 255, 255))
    image_masked_edges = cv2.bitwise_and(image_input, mask)
    
    # Detect color YELLOW
    image_hsv = cv2.cvtColor(image_masked_edges,cv2.COLOR_BGR2HSV)
    image_yellow = cv2.inRange(image_hsv, (18,110,110), (25,255,255)) 
    
   
    # Detect color WHITE
    image_white = np.copy(image_masked_edges)
    print(image_white)
    print(image_yellow)
    #white_pixels = (image_white[:,:,0] < 212) \
     #       | (image_white[:,:,1] < 212) \
    #        | (image_white[:,:,2] < 212)
    #image_white[white_pixels] = [0,0,0]

    #image_white = cv2.inRange(image_masked_edges,(0,0,212), (0,0,255)) 

    #result = cv2.bitwise_or(image_yellow, image_white)
    return image_white




#Listing images 
list_images_names = os.listdir("test_images/")

for img_name in list_images_names:
    img_input = cv2.imread(r'test_images/'+img_name)
    img_output = process_image(img_input)
    plt.imshow(img_output[...,::-1])
    plt.show()
    cv2.imwrite(r'test_images_output/Output4_'+img_name, img_output)
    
```
